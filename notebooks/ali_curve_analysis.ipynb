# Argon 200 ms, 1bar, 50 pulses 

```python
dfAr = load_raw_ali_df('../ali_data/31052019_Ar_1010mbar_200ms_50pulses')
```

```python
pAr, tAr = plot_pressure_curve(dfAr, tit='Argon curves')
```

```python
len(pAr), len(tAr)
```

```python
overlay_peaks(dfAr, pAr, tAr, npulses=50, xrange=200)
```

```python
xrange = 200
matrix = [dfAr.p_chamber[tAr[i]:tAr[i]+xrange].values for i in range(50)]
dfpeakAr = pd.DataFrame(list(map(np.ravel, matrix))).transpose()
av_p, sd_p = np.zeros(xrange), np.zeros(xrange)
av_p = dfpeakAr.mean(axis=1)
sd_p = dfpeakAr.std(axis=1)
```

```python
overlay_peaks(dfAr, pAr, tAr, npulses=10, xrange=200, scale='ms')
plt.plot(dfpeakAr[:100].index, av_p[:100], label = 'average curve')
plt.fill_between(dfpeakAr[:100].index, (av_p[:100]-sd_p[:100]), (av_p[:100]+sd_p[:100]), color='0.9', label = 'statistical uncertainty')
plt.legend(loc='best')
```
